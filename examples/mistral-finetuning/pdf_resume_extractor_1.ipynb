```python
def group_keys_by_nearest(keys, n):
    keys.sort()  # Sort the keys in ascending order
    result = [[]]  # Initialize the result list with an empty group
    current_group = result[0]  # Start with the first group

    for key in keys:
        if not current_group:
            current_group.append(key)
        else:
            diff = key - current_group[-1]

            if diff <= n:
                current_group.append(key)
            else:
                result.append([key])
                current_group = result[-1]

    return result
```
