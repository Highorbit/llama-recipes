```python
import pandas as pd
import numpy as np
from tqdm.notebook import tqdm
import pickle
```

```python
user_data = pd.read_csv('custom_data/validation_dataset.csv')
```

```python
user_data.sample(5)
```

```python
pi_validation = []

for row in tqdm(user_data.itertuples()):
    try :
        feat = {
            'phone_number' : int(row.phone),
            'email_id' : row.email,
            'name' : row.name,
            'id': row.id 
        }
    except:
        pass
    
    pi_validation.append(feat)
```

```python
pi_validation[:3]
```

```python
with open('custom_data/validation_output.pkl','rb') as f:
    v_data = pickle.load(f)
```

```python
v_data[:2]
```

```python
v_pred = {}
for data in tqdm(v_data):
    key = list(data.keys())[0]
    val = list(data.values())[0]
    v_pred[key] = val
```

```python
v_truth = {}
for data in tqdm(pi_validation):
    key = data['id']
    val = data
    v_truth[key] = val
```

```python
email_count = 0
phone_count = 0
name_count = 0
error_count = 0
error_list = []
name_list = []
email_list = []

for userid in tqdm(list(v_pred.keys())):
    
    try:
        pred = v_pred[userid]
        truth = v_truth[userid]
        
        if int(pred['phone'][-10:])==truth['phone_number']:
            phone_count+=1
            
        else:
            p = int(pred['phone'][-10:])
            t = truth['phone_number']
            
            if len(p.split(','))>1:
                pc = p.split(',')[0]
                
                if pc==truth['phone_number']:
                    phone_count+=1
            
            print(f'User ID {userid} Predicted {p} Truth {t}')
        
        if pred['email'].lower()==truth['email_id'].lower():
            email_count+=1
        else:
            email_list.append(userid)

        if pred['name'].lower()==truth['name'].lower():
            name_count+=1
        else:
            name_list.append(userid)
            
    except:
        error_count+=1
        error_list.append(userid)
```

```python
den = len(v_pred.keys())-error_count
```
