```python
import json
import ast
import re
```

```python
with open('data/test_results.json','r') as f:
    data = json.load(f)
```

```python
converted = {}
errors = []
for key, value in data.items():
    try:
        out_json = ast.literal_eval(value)
        converted[key] = out_json
    except:
        errors.append(key)
        pass
```
